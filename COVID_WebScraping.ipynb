```python
from IPython.core.display import display, HTML
import requests
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
covid_url = "https://web.archive.org/web/20210903052746/https://dph.illinois.gov/news/202004"
covid_response = requests.get(covid_url)
```

```python
covid_response # successful request(promise resolved)
print(covid_response.text) # returns page src(markup, styles, script)
```

```python
soup = BeautifulSoup(covid_response.text, 'html.parser')
print(soup)
```

```python
import re
day_link = soup.find('a', attrs={'href': re.compile('202004')})
```

```python
target_date_str = '29th Apr, 2020'

target_div = None
for div in soup.find_all('div', {'class', 'views-field views-field-created'}):
    span_text = div.span.get_text()
    if span_text == target_date_str: 
        print ("Found target!")
        target_div = div
        print(target_div)
```

```python
IDPH_BASE = 'https://web.archive.org'
detail_link = IDPH_BASE + target_div.parent.find('span', {'class': 'field-content'}).a.get('href')
print (detail_link)
```

```python
detail_req = requests.get(detail_link)
detail_soup = BeautifulSoup(detail_req.content, 'html.parser')
print (detail_soup)
```

```python
news_div = detail_soup.find('div', {'class': 'field-item even'})

county_list = []
entries_list = []

for li in news_div.ul.find_all('li'):
    county, text = li.text.split('County')
    entries = text[2:].split(',')
    print (county)
    print (entries)
    county_list.append(county)
    entries_list.append(entries)
```

```python
print (county_list)
print (entries_list)
```

```python
def extract(entry):
    sl = entry.split(' ')
    sl = [s for s in sl if s]
    count = int(sl[0])
    sex = sl[1]
    age = int(sl[2][:len(sl)-1])
    return count, sex, age

counties, counts, sexes, ages = [], [], [], []
for county, entries in zip(county_list, entries_list):
    for entry in entries:
        if entry != '':
          count, sex, age = extract(entry)  # entry is a string like "2 male 50s"
          counties.append(county)
          counts.append(count)
          sexes.append(sex)
          ages.append(age)
        
out = pd.DataFrame(data={'County': counties, 'Count': counts, 'Sex': sexes, 'Age_bracket': ages})
print(out) 
```
